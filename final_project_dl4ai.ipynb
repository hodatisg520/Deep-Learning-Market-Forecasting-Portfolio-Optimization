## Task 1.1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout

# Đọc dữ liệu
df = pd.read_csv('AAL.csv')

df.head()
```

```python
# Get some info about the data
df.info()
```

```python
# Chọn 6 đặc trưng theo yêu cầu đề bài
features = ['Low', 'High', 'Open', 'Close', 'Adjusted Close', 'Volume']
dataset = df[features].values

# 2. Chuẩn hóa dữ liệu (Scaling)
# Dùng 2 scaler riêng biệt để dễ dàng khôi phục giá trị dự đoán về đơn vị gốc
scaler_features = MinMaxScaler(feature_range=(0, 1))
scaler_target = MinMaxScaler(feature_range=(0, 1))

scaled_features = scaler_features.fit_transform(dataset)
# Giả sử ta muốn dự đoán giá 'Close' (cột index 3)
target_col_idx = 3
scaler_target.fit(dataset[:, target_col_idx].reshape(-1, 1))
```

```python
def create_dataset(data, target_idx, time_step=60):
    X, y = [], []
    for i in range(time_step, len(data)):
        # X chứa 6 đặc trưng của 'time_step' ngày trước đó
        X.append(data[i-time_step:i, :])
        # y là giá mục tiêu của ngày hiện tại
        y.append(data[i, target_idx])
    return np.array(X), np.array(y)

time_step = 60 # Sử dụng 60 ngày quá khứ để dự đoán [cite: 817]
X, y = create_dataset(scaled_features, target_col_idx, time_step)

# Chia tập Train/Test theo trình tự thời gian
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
model = Sequential([
    # Input shape: (số bước thời gian, số đặc trưng) -> (60, 6)
    LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2), # Chống overfitting [cite: 1012]

    LSTM(units=50, return_sequences=False),
    Dropout(0.2),

    Dense(units=25),
    Dense(units=1) # Dự đoán 1 giá trị đầu ra
])

model.compile(optimizer='adam', loss='mean_squared_error')

# Huấn luyện mô hình
model.fit(X_train, y_train, batch_size=32, epochs=50, validation_data=(X_test, y_test))
```

```python
# Dự đoán trên tập test
predictions = model.predict(X_test)

# Chuyển ngược về giá trị thực tế
predictions_rescaled = scaler_target.inverse_transform(predictions)
y_test_rescaled = scaler_target.inverse_transform(y_test.reshape(-1, 1))

# Vẽ biểu đồ so sánh [cite: 11]
plt.figure(figsize=(12,6))
plt.plot(y_test_rescaled, color='blue', label='Giá thực tế')
plt.plot(predictions_rescaled, color='red', label='Giá dự đoán')
plt.title('Dự đoán giá cổ phiếu Nasdaq (Đa biến)')
plt.xlabel('Thời gian')
plt.ylabel('Giá')
plt.legend()
plt.show()
```

## Task 1.2

```python
import numpy as np

def create_dataset_kth_day(data, target_idx, time_step=60, k=3):
    """
    Hàm tạo dữ liệu dự đoán ngày thứ k.
    k: số ngày dự đoán trước (VD: 3 hoặc 7)
    """
    X, y = [], []
    # Lưu ý vòng lặp dừng lại sớm hơn để có đủ dữ liệu cho ngày thứ k
    for i in range(len(data) - time_step - k + 1):
        # X lấy 'time_step' ngày (ví dụ 60 ngày)
        X.append(data[i:(i + time_step), :])

        # y là giá trị ở 'time_step + k - 1' ngày tính từ i
        y.append(data[i + time_step + k - 1, target_idx])

    return np.array(X), np.array(y)
```

```python
# Giả sử 'scaled_features' và 'target_col_idx' đã được định nghĩa như ở Task 1.1

time_step = 60
k_days_ahead = 3 # Hoặc k=7 theo yêu cầu

# Tạo X, y mới
X_k, y_k = create_dataset_kth_day(scaled_features, target_col_idx, time_step, k=k_days_ahead)

# Chia Train/Test (Luôn giữ thứ tự thời gian)
train_size = int(len(X_k) * 0.8)
X_train_k, X_test_k = X_k[:train_size], X_k[train_size:]
y_train_k, y_test_k = y_k[:train_size], y_k[train_size:]
```

```python
# Khởi tạo mô hình
model_k = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_train_k.shape[1], X_train_k.shape[2])),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25),
    Dense(1) # Vẫn dự đoán 1 giá trị (cho ngày thứ k)
])

model_k.compile(optimizer='adam', loss='mean_squared_error')

# Huấn luyện
history_k = model_k.fit(X_train_k, y_train_k, batch_size=32, epochs=50, validation_data=(X_test_k, y_test_k))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
# 1. Dự đoán
predictions_k_scaled = model_k.predict(X_test_k)

# 2. Khôi phục giá trị thực
predictions_k_real = scaler_target.inverse_transform(predictions_k_scaled)
y_test_k_real = scaler_target.inverse_transform(y_test_k.reshape(-1, 1))

# 3. Tính toán các chỉ số đánh giá lỗi (MAE và RMSE)
mae_k = mean_absolute_error(y_test_k_real, predictions_k_real)
rmse_k = np.sqrt(mean_squared_error(y_test_k_real, predictions_k_real))

print(f"Mean Absolute Error (MAE) cho ngày thứ {k_days_ahead}: {mae_k:.4f}")
print(f"Root Mean Squared Error (RMSE) cho ngày thứ {k_days_ahead}: {rmse_k:.4f}")

# 4. Vẽ đồ thị trực quan hóa kết quả
plt.figure(figsize=(14, 6))
plt.plot(y_test_k_real, color='blue', label='Giá thực tế (Actual)')
# Dùng alpha=0.7 để làm mờ đường dự đoán một chút, giúp dễ nhìn khi hai đường chồng lên nhau
plt.plot(predictions_k_real, color='red', alpha=0.7, label=f'Giá dự đoán (Predicted - {k_days_ahead} days ahead)')

plt.title(f'So sánh Giá thực tế và Giá dự đoán (Ngày thứ {k_days_ahead})')
plt.xlabel('Thời gian (Các ngày trong tập Test)')
plt.ylabel('Giá cổ phiếu')
plt.legend()
plt.grid(True)
plt.show()
```
